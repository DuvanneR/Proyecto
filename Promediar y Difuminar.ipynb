```python
import cv2 #Libreria de OpenCV.
import numpy as np #Libreria Numerica de Python.
import matplotlib.pyplot as plt #Libreria para mostrar Graficos o imagenes.
from PIL import Image
```

```python
I = cv2.imread('D.JPG') #Subir imagen en formato BGR.
Iv2 = cv2.cvtColor(I, cv2.COLOR_BGR2RGB) #Cambiar formato de la imagen de BGR a RGB.
```

```python
plt.imshow(Iv2) #Mostrar imagen en formato RGB
```

```python
Promediar = np.ones((6,6),np.float32)/25 #Promedia los pixeles de la imagen para modificar el ruido de la imagen.
 
Iv3 = cv2.filter2D(Iv2,-1,Promediar) #Agrega el promedio de pixeles anterior a la imagen.

plt.imshow(Iv3) #Mostrar imagen con un ruido modificado.
```

```python
Difuminar = cv2.blur(Iv2,(50,50)) #Difumina los pixeles de la imagen.
 
plt.imshow(Difuminar) #Mostrar imagen con difuminado.
```
